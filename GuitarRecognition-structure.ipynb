```python
g = 9.81 # m/s^2
B = 80 # mT

r = np.linspace(0,10,100) # cm
delta = np.linspace(0,np.pi/2, 5)
```

### Eksmpel på kode og resultater + noen tips

Denne funksjonen er en eksempel-funksjon som brukes til å lage et eksempel-plot.

```python
def example_func(r, delta):
    return g*2*B*np.exp(1j*r+delta)
```

Her er et plott med en harmonisk svingebevegelse. (Husk å ha navn på aksene og enheter!)

```python
plt.figure(figsize=(8,6))

for i in range(len(delta)):
    plt.plot(r, np.real(example_func(r, delta[i])), label="$\delta$ = %.2f"%delta[i])
    
plt.legend()
plt.title("Harmonisk svingebevegelse")
plt.xlabel("r [cm]")
plt.ylabel("amplitude [enhet]")
```

Det er også viktig å forklare leseren hva du har lagt merke til i plottet. For eksempel: Figuren over viser at amplituden blir større når delta øker. Det er nok fordi å addere $b$ i potensen til $e$ er det samme som å gange med $e^b$: $e^{a + b} = e^{a}e^{b}$.

For å føre essayet fremover kan du etter å ha undersøkt den enkle modellen, og eventuelt fått et svar på problemstillingen for den, øke kompleksiteten av modellen. Da kan du introdusere de nye formelene og fysikken, og så gjøre det samme med den nye mer kompliserte modellen.
